```python
# this is some code that just visulaises a models feature importance
# useful for comparing your new trained models to the baseline
```

```python
import joblib
import numpy as np
import matplotlib.pyplot as plt

model = joblib.load("logreg_model.joblib")
features = ["tag_h1_coordination_game", "tag_h2_status_seekers", 
            "tag_h3_shared_interests", "tag_h4_profile_effort", 
            "tag_h5_likely_female"]


coefficients = model.coef_[0]

sorted_idx = np.argsort(np.abs(coefficients))[::-1]

plt.barh([features[i] for i in sorted_idx], [coefficients[i] for i in sorted_idx])
plt.xlabel("Coefficient Value")
plt.title("Feature Importance (Logistic Regression)")
plt.gca().invert_yaxis()
plt.show()
```
